```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/content/house_train.xlsx"
df = pd.read_excel(file_path)
```

Check for missing values in the dataset. Which columns have missing values, and how many
missing values are there in each of these columns?

```python
print("Missing values in dataset:")
print(df.isnull().sum())
```

Handle the missing values in the dataset. For categorical columns, fill missing values with
appropriate method.

```python
for col in df.columns:
    if df[col].dtype == "object":  # Categorical columns
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:  # Numerical columns
        df[col].fillna(df[col].median(), inplace=True)
```

Verify that there are no remaining missing values after this operation.

```python
print("\nMissing values after handling:", df.isnull().sum().sum())
```

Remove a specific row from the dataset. Choose a row to remove based on the index (e.g.,
remove the first row). After removing the row, display the shape of the DataFrame to show that
the row has been removed.

```python
df = df.drop(index=0).reset_index(drop=True)
print("\nShape after removing first row:", df.shape)
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure df is properly defined
numeric_df = df.select_dtypes(include=[np.number])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(22, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

Feature selection using correlation

```python
import pandas as pd

# Select only numerical features
numeric_df = df.select_dtypes(include=["number"])

# Compute correlation with SalePrice
correlation_with_target = numeric_df.corr()["SalePrice"].sort_values(ascending=False)

# Set threshold for strong correlation (above 0.5 or below -0.5)
threshold = 0.5
selected_features = correlation_with_target[correlation_with_target.abs() > threshold]

# Display selected features
print("Selected Features with High Correlation to SalePrice:\n")
print(selected_features)
```
