```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "/content/house_train.csv"
df = pd.read_csv(path)
```

```python
(df.isnull().sum())
```

```python
for col in df.columns:
    if df[col].dtype == "object":  # Categorical
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:  # Numerical
        df[col].fillna(df[col].median(), inplace=True)
```

```python
df.head()
```

```python
#missing values
df.isnull().sum().sum()
```

```python
df = df.drop(index=0).reset_index(drop=True)
df.shape #after removing 1st row
```

```python
if 'Id' in df.columns:
    df.drop(columns=['Id'], inplace=True)
df.columns
```

```python
df['Price.Per.Sqft'] = df['SalePrice'] / df['LotArea']
print("\nNew column 'Price.Per.Sqft' added:")
print(df[['Price.Per.Sqft']].head())
```
